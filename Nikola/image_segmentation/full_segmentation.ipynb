```python
# Load images in colored and not colored manner
img_color = cv2.imread('../images_of_india/final_map.png')
#img_grayscale = cv2.imread('../images_of_india/final_map.png',0)
```

```python
import matplotlib
import matplotlib.pyplot as plt
# import geopandas as gpd
# import plotly.express as px
import pandas as pd
```

```python
stations_path = ("./datasets_with_coordinates/dataset_2_ANFIS_with_long_lat.csv")
```

```python
df = pd.read_csv(stations_path)
```

```python
df
```

```python
#image information
#This information is gotten from: 

#longitude is vertical.. However, longitudes change horizontaly!
min_latitude = 8.0590000
max_latitude = 35.8530000

min_longitude = 67.9390000
max_longitude = 97.3830000
```

```python
#calculating the amount of maximum longitude and latitude on the image
real_latitude_span = max_latitude - min_latitude
real_longitude_span = max_longitude - min_longitude

print(real_latitude_span)
print(real_longitude_span)
```

```python
# The amount of pixels of the final image.
max_x_pixel_lat = 672
max_y_pixel_long = 692
```

```python
#Create a dictionary with all station codes, latitude and longitude
my_list= [(x,y,z)for x,y,z in df[['WQI_pred','latitude','longitude']].itertuples(index = False, name= None)]
my_list
```

```python
#the amount of latitude one pixel covers
one_pixel_latitude = real_latitude_span/max_x_pixel_lat

#the amount of longitude one pixel covers
one_pixel_longitude = real_longitude_span/max_y_pixel_long

```

```python
#taking the WQI latitude and longitude from the dataset
for WQI, latitude, longitude in my_list:
    print(WQI, latitude, longitude)
    
    #reset according to pixels 0,0
    # subtracting the minimum latitude and longitude on the image
    latitude = latitude - min_latitude
    longitude = longitude - min_longitude
    
    #find what pixels need to color
    # how many pixels from the origin we need to place the square
    lat = latitude / one_pixel_latitude
    long = longitude /one_pixel_longitude
    
    #Converting the values to ints... We could possibly disregard this??
    par1 = int(lat)
    par2 = int(long)
    
    #Assigning different colors based on the WQI classification
    if WQI <= 25:
        img_color[(692-par1):(692-par1+5),par2:(par2+5)] = (0,255,0)
    elif WQI <= 50: 
        img_color[(692-par1):(692-par1+5),par2:(par2+5)] = (255,128,0)
    elif WQI <= 75:
        img_color[(692-par1):(692-par1+5),par2:(par2+5)] = (0,128,255)
    else:
        img_color[(692-par1):(692-par1+5),par2:(par2+5)] = (0,0,255)
    
```
